```python
import pandas as pd
import pycountry
```

```python
df = pd.read_csv('archive (1)/states_all.csv')
```

```python
df.shape
```

```python
df.columns
```

```python
df.STATE.unique()
```

```python
df = df.loc[~df['STATE'].isin(['DISTRICT_OF_COLUMBIA', 'DODEA', 'NATIONAL'])]
```

```python
df.YEAR.unique()
```

```python
df = df[df['YEAR']==2017]
```

```python
df.shape
```

```python
df = df.drop(['PRIMARY_KEY','YEAR'], axis=1)
```

```python
df = df.dropna(axis=1)
```

```python
df.columns
```

```python
df = df[['STATE', 'AVG_MATH_8_SCORE', 'AVG_READING_8_SCORE']]
```

```python
# Define a dictionary to map the old column names to the new column names
new_names = {
    'STATE': 'state',
    'AVG_MATH_8_SCORE': 'math_score',
    'AVG_READING_8_SCORE': 'reading_score'
}

# Rename the columns using the dictionary
df = df.rename(columns=new_names)
```

```python
df['state'] = df['state'].str.replace('_', ' ')
```

```python

# Create a dictionary of state names and abbreviations
states_dict = {subdivision.name.upper(): subdivision.code.split('-')[1] for subdivision in pycountry.subdivisions if subdivision.country_code == 'US'}

# Replace state names with abbreviations in the 'state' column
df['state'] = df['state'].replace(states_dict)
```

```python
df = df.reset_index(drop=True)
df.index += 1
df
```
